```python
import datetime
from datetime import date
"""
Converting the input file into a dictionary with values y, x, 
"""
def convert_input(delivery):
    for key in delivery:
        s = delivery[key]
        s = s.split(" ")
        i = 0
        lister = []
        p = s[0]
        t = list(p)
        for k in range(len(t)):
            if not t[k].isdigit():
                temp = 0
                if t[k] == "S":
                    iter = k+1
                    while iter < len(t) and t[iter].isdigit():
                        temp = temp*10 + int(t[iter])
                        iter+=1
                    lister.append(-(temp))
                elif t[k] == "N":
                    iter = k +1
                    while iter < len(t) and t[iter].isdigit():
                        temp = temp*10 + int(t[iter])
                        iter+=1
                    lister.append(temp)
                elif t[k] == "E":
                    iter = k +1
                    while iter < len(t) and t[iter].isdigit():
                        temp = temp*10 + int(t[iter])
                        iter+=1
                    lister.append(temp)
                elif t[k] == "W":
                    iter = k+1
                    while iter < len(t) and t[iter].isdigit():
                        temp = temp*10 + int(t[iter])
                        iter+=1
                    lister.append(-(temp))
                else:
                    continue
        k = s[1]
        k = k.split(":")
        lister.append(datetime.time(int(k[0]),int(k[1]),int(k[2])))
        lister.append(datetime.time(int(k[0])+1,int(k[1]),int(k[2])))
        #print(lister)
        delivery[key] = lister
        
    return delivery
```

```python

"""
Calculate the distance 

"""
def dist(y, x):
    d = abs(y) + abs(x)
    return d
```

```python
#import distance_calculator
"""
get the Manhattan distance for every order

"""
def delivery_distance(delv_schedule):
    
    for key in delv_schedule:
        dval = dist(delv_schedule[key][0], delv_schedule[key][1])
        delv_schedule[key].append(dval)
        
    return delv_schedule
    

#delivery_schedule
```

```python
def calc_priority(delivery_schedule):
    
    priority = []
    d_start_time = dt.time(6,0,0)

    for key in delivery_schedule:
        #If order placement time > start time, the delivery time will be calculate from the start of the delivery time 
        if dt.datetime.combine(dt.date(1,1,1),delivery_schedule[key][2]).time() >= d_start_time:
            delivery_time = (dt.datetime.combine(dt.date(1,1,1),delivery_schedule[key][2]) + dt.timedelta(minutes = delivery_schedule[key][4])).time()
        else: 
            #The delivery time will be calculated from the start time
            delivery_time = (dt.datetime.combine(dt.date(1,1,1),d_start_time) + dt.timedelta(minutes = delivery_schedule[key][4])).time()
    #         d_start_time = (dt.datetime.combine(dt.date(1,1,1),d_start_time) + dt.timedelta(minutes = 2*delivery_schedule[key][4])).time()
        #if delivery time is less than one hour calculated time, we add it to the prority array
        if delivery_time <= delivery_schedule[key][3]:
            priority.append(key)
    #For keys not in priority we add it to schedule 
    for key in delivery_schedule:
        if dt.datetime.combine(dt.date(1,1,1),delivery_schedule[key][2]).time() >= d_start_time:
            delivery_time = (dt.datetime.combine(dt.date(1,1,1),delivery_schedule[key][2]) + dt.timedelta(minutes = delivery_schedule[key][4])).time()
        else: 
            #The delivery time will be calculated from the start time
            delivery_time = (dt.datetime.combine(dt.date(1,1,1),d_start_time) + dt.timedelta(minutes = delivery_schedule[key][4])).time()
        if dt.time(1,0,0)< delivery_time <= dt.time(2,0,0):
            priority.append(key)
    for key in delivery_schedule:
        if key not in priority:
            priority.append(key)
    # sorted_distances = sorted(delivery_schedule, key = lambda k: (delivery_schedule[k][3]))
    # print(sorted_distances)
    return priority
```

```python
#import priority_calculator as pc

import datetime as dt
def delivery_order(delivery_schedule, start, end):
    priority = calc_priority(delivery_schedule) #pc.priority(delivery_schedule)
    next_time = start
    end_time = end
    
    #next_time = dt.time(6,0,0)
    #end_time = dt.time(22,0,0)
    if next_time <= end_time:
        for key in priority:
            #start time till be the order placement time if the start time is less than the order placement time 
            if next_time < delivery_schedule[key][2]:
                start_time = delivery_schedule[key][2]
            else: start_time = next_time


            delivery_time = (dt.datetime.combine(dt.date(1,1,1),start_time) + dt.timedelta(minutes = delivery_schedule[key][4])).time()
            #print("Delivery Time", delivery_time)
            delivery_schedule[key].append(delivery_time)
        #if the start time is before recieved time 
            next_time = (dt.datetime.combine(dt.date(1,1,1),start_time) + dt.timedelta(minutes = 2 * delivery_schedule[key][4])).time()
            
    return delivery_schedule
```

```python
from datetime import datetime as dtime
from datetime import timedelta as td
from datetime import date as dte
import datetime as dt1

"""
NPS Calculation

"""

def nps(deliv, pcount, dcount):
    
    totalc = len(deliv)
    for key in deliv:
        duration = dtime.combine(date.min, deliv[key][5]) - dtime.combine(date.min, deliv[key][2])
        #print(duration)
        #print(dt.time(1,0,0))
        if duration <= dt1.timedelta(hours = 1, minutes = 0, seconds = 0):
            pcount += 1
        if duration > dt1.timedelta(hours = 4, minutes = 0, seconds = 0):
            dcount += 1
    npscal = (pcount/totalc * 100) - (dcount/totalc * 100)
    
    return npscal
            
```
